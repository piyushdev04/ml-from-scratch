# Linear Regression From Scratch

#### Step 1 → Generate sample data

```python
import numpy as np

np.random.seed(42)
X = np.random.rand(100, 1) * 10 # 100 points between 0 - 100
true_m = 2.5
true_c = 7
noise = np.random.randn(100, 1)

y = true_m * X + true_c + noise
```

#### Step 2 → Initialize parameters

```python
m = np.random.randn()
c = np.random.randn()
lr = 0.01 # learning rate
```

#### Step 3 → Define model + loss

```python
def predict(X):
    return m * X + c

def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)
```

#### Step 4 → Gradient Descent Update Rule

```python
for epoch in range(1000):
    y_pred = predict(X)

    # gradients
    dm = -2 * np.mean(X * (y - y_pred))
    dc = -2 * np.mean(y - y_pred)

    # update params
    m -= lr * dm
    c -= lr * dc

    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss = {mse_loss(y, y_pred):.4f}")
```

```python
print("Learned m:", m)
print("Learned c:", c)
```

```python
# graph
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

# Plot data points
plt.scatter(X, y, alpha=0.7, label='Actual Data')

# Plot learned line
plt.plot(X, predict(X), linewidth=3, label='Predicted Line')

# Labels & Title
plt.xlabel("X (Input Feature)")
plt.ylabel("Y (Target Value)")
plt.title("Linear Regression Fit")

# Visual improvements
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
